# Dati di misura
I dati di misura sono stati ricavati da due esperimenti realizzati rispettivamente nelle date:
- Agosto 1954 - Novembre 1955
- Settembre 1957 - Dicembre 1957
Entrambi gli esperimenti sono stati effettuati nella stessa zona (Round Hill Field Station, Massachusetts Institute of Technology) con un setup simile.
Dettagli:
## Round Hill 1
### Emissione del gas
- Fonte di SO2 situata ad un'altezza di 30 cm dal livello del terreno
### Misure di concentrazione
- Le misure sono state effettuate in 3 semicirconferenze di raggio 50, 100, 200 m, con centro nella fonte del gas, tramite sensori posti lungo la circonferenza ad una distanza di 1.5 gradi l'uno dall'altro. Ogni sensore misura la concentrazione media di SO2 nell'aria per un periodo di 10 minuti

## Round Hill 2
### Emissione del gas
- Fonte di SO2 situata ad un'altezza di 1.5 m dal livello del terreno
### Misure di concentrazione
- Le misure sono state effettuate ad un'altezza di 1.5 m in 3 semicirconferenze di raggio 50, 100, 200 m con centro nella fonte del gas. In questo caso i sensori 1-6 e 106-111 sono distanziati di 3 gradi, mentre gli altri di 1.5 gradi. Le misure di concentrazione media sono state effettuate per un periodo di 10, 3, 1.5 min (Tutti gli intervalli di misurazione iniziano nello stesso istante)

### Misure meteorologiche (Round Hill 1 e Round Hill 2)
Durante l'esperimento sono state effettuate le seguenti misurazioni, in un periodo di 20 minuti centrato nei 10 minuti di misura:
- Velocità del vento media
- Temperatura
Questi due parametri sono stati misurati a diverse altezze (1.5, 3, 6 e 12 m)
- Distribuzione della direzione del vento
Quest'ultima è stata misurata ad un'altezza di 2 m

## Correzioni
I due dataset presentano dei fattori di correzione da applicare alle misure di concentrazione. Questi fattori moltiplicativi tengono conto dell'evaporazione della fonte di SO2, avvenuta più velocemente del previsto a causa delle temperature, portando quindi a misure di concentrazione più elevate

# Interpretazione dei dati


### data1 - data2
Entrambi i dataset organizzati in un formato descrittivo, non adatto a processi di data analysis, cleaning o imputation.

Entrambi i dataset sono divisi in tabelle che contengono:
- Header (giorno, mese, anno, iterazione (run), intervallo orario)
- relhgt (altezza della fonte di SO2 dal terreno)

Colonne:
- Post: identificativo del sensore (intero)
- Degrees: angolo di inclinazione del sensore rispetto al sensore #1 (gradi)
- Azimuth: angolo di inclinazione del sensore rispetto al Nord (gradi)
- Wind-Dir: rappresenta quanto spesso la velocità del vento misurata è osservata al sensore. Questa misura è rappresentata in percentuale. La somma delle percentuali è posta in fondo alla rispettiva colonna (Non eguaglia sempre il 100%)
- Concentrazioni: colonna che racchiude 3 misure di concentrazione in $mg/m^3$ per le tre distanze considerate: 50, 100, 200 m

### Round Hill 2
Il secondo dataset è strutturato nello stesso modo. Le misure di concentrazione sono, questa volta, 9 in totale, 3 per ogni intervallo di tempo (10 min, 3 min, 1.5 min).<br>*La colonna relativa alla misurazione di concentrazione a 1.5 min è indicata con con "0.5-min", questo è probabilmente un errore avvenuto durante la trascrizione del dataset cartaceo*.<br>
Sono inoltre presenti misurazioni verticali, rispetto ai sensori, della concentrazione, rispettivamente a 2.5, 1.0, 0.5 m di altezza.



### meteo1 - meteo2
I due fogli contengono dati meteorologici registrati durante le misurazioni di concentrazione del gas.
Le colonne rappresentano:
- U (m/s) : velocità del vento orizzontale alla fonte
- Sa (deg) : deviazione standard dell'angolo di elevazione del vento misurato
- Se (deg) - Non presente in meteo2 : deviazione standard dell'angolo orizzontale del vento misurato
- U1.5, U3.0, U6.0, U12.0 (m/s) : velocità orizzontale del vento misurata ad altezze diverse (1.5 m, 3.0 m, 6.0 m, 12.0 m) da torri di osservazione situate nelle vicinanze della fonte del gas
- T1.5, T3.0, T6.0, T12.0 (C) : temperatura misurata dalle torri alle stesse altezze

Valori non rilevati vengono rappresentati da -99

### correction1 - correction2
I due fogli contengono i fattori di correzione moltiplicativi che tengono conto del tasso di evaporazione del gas. I valori sono compresi tra 0 e 1.0, riducendo la concentrazione effettiva del gas.
La colonna Q rappresenta il tasso di evaporazione del gas in g/s
